```python
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import NearestNeighbors
CSV_PATH = "liver_cirrhosis.csv"
LABEL_COL = "Stage"

# 유사 판정 기준:
# cosine 유사도 기반으로 하려면 -> metric="cosine", threshold는 0.999~0.9999 권장
# euclidean 기반으로 하려면 -> metric="euclidean", threshold는 스케일에 따라 조정 필요
METRIC = "cosine"
SIM_THRESHOLD = 0.9995  # (cosine) 1에 가까울수록 더 빡셈. 처음엔 0.999~0.9997 권장

# =========================
# 1) 로드
# =========================
df = pd.read_csv(CSV_PATH)

# 라벨 없는 경우 방지
assert LABEL_COL in df.columns, f"'{LABEL_COL}' 컬럼이 없습니다."

# feature / label 분리
X_raw = df.drop(columns=[LABEL_COL])
y = df[LABEL_COL].copy()

# =========================
# 2) 완전 중복 제거 (전체 컬럼 동일)
# =========================
before = len(df)
df_no_exact = df.drop_duplicates(keep="first").reset_index(drop=True)
print(f"[Exact dup] 제거: {before - len(df_no_exact)} / 남음: {len(df_no_exact)}")
```

```python
# =========================
# 3) feature 중복 그룹 + 라벨 충돌 리포트
# =========================
feat_cols = [c for c in df_no_exact.columns if c != LABEL_COL]

# feature가 동일한 행들을 그룹으로 묶음
g = df_no_exact.groupby(feat_cols, dropna=False)[LABEL_COL]

# feature 동일 그룹 중, 라벨이 2개 이상인 "충돌" 그룹
conflict_keys = g.nunique()
conflict_keys = conflict_keys[conflict_keys > 1]
print(f"[Label conflict] feature 동일인데 라벨 다른 그룹 수: {len(conflict_keys)}")

# 충돌 그룹 샘플 일부 저장(검토용)
if len(conflict_keys) > 0:
    conflict_df = (
        df_no_exact.merge(
            conflict_keys.rename("n_labels").reset_index()[feat_cols],
            on=feat_cols,
            how="inner"
        )
        .sort_values(feat_cols + [LABEL_COL])
    )
    conflict_df.head(200).to_csv("conflict_feature_duplicates_sample.csv", index=False)
    print(" - conflict 샘플 저장: conflict_feature_duplicates_sample.csv (최대 200행)")

# 여기서 정책 선택:
# (A) 충돌 그룹은 제거 (보수적, 추천)
# (B) 다수결로 라벨 통일
REMOVE_CONFLICT_GROUPS = True

if REMOVE_CONFLICT_GROUPS and len(conflict_keys) > 0:
    # conflict group key만 추출
    conflict_key_df = conflict_keys.reset_index()[feat_cols]
    df_no_conflict = df_no_exact.merge(
        conflict_key_df.assign(_conflict=1),
        on=feat_cols,
        how="left"
    )
    df_no_conflict = df_no_conflict[df_no_conflict["_conflict"].isna()].drop(columns=["_conflict"])
    print(f"[Remove conflicts] 제거: {len(df_no_exact) - len(df_no_conflict)} / 남음: {len(df_no_conflict)}")
else:
    df_no_conflict = df_no_exact
```

```python
# =========================
# 4) near-duplicate 제거를 위한 전처리 (one-hot + scaling)
# =========================
X = df_no_conflict.drop(columns=[LABEL_COL])
y = df_no_conflict[LABEL_COL].copy()

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X.columns if c not in num_cols]

preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(with_mean=True, with_std=True), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ],
    remainder="drop",
)

# 변환
X_vec = preprocess.fit_transform(X)

# sparse일 수 있으니 NearestNeighbors가 처리 가능
# metric="cosine"이면 거리 = 1 - cosine_similarity
nn = NearestNeighbors(n_neighbors=2, metric=METRIC, algorithm="auto")
nn.fit(X_vec)

dist, idx = nn.kneighbors(X_vec, return_distance=True)
# dist[:,0]는 자기 자신 거리(0), dist[:,1]이 가장 가까운 이웃 거리
nearest_dist = dist[:, 1]
nearest_idx = idx[:, 1]

if METRIC == "cosine":
    # cosine distance -> similarity 변환
    nearest_sim = 1 - nearest_dist
    too_close = nearest_sim >= SIM_THRESHOLD
else:
    # euclidean이면 dist가 작을수록 가까움 -> threshold는 사용자가 조정
    # 여기서는 SIM_THRESHOLD를 "거리 임계값"으로 해석
    too_close = nearest_dist <= SIM_THRESHOLD

print(f"[Near-dup] 기준 충족(너무 가까움) 샘플 수(플래그): {too_close.sum()}")

# =========================
# 5) near-duplicate 그룹에서 1개만 남기기
#    - 간단한 방식: "나와 가장 가까운 이웃이 너무 가깝다"면,
#      더 작은 인덱스만 남기고 나머지는 제거
# =========================
to_drop = set()
for i in range(len(X)):
    if not too_close[i]:
        continue
    j = int(nearest_idx[i])
    # i와 j가 너무 비슷: 둘 중 하나 제거 (여기서는 큰 인덱스를 제거)
    drop_one = max(i, j)
    to_drop.add(drop_one)

keep_mask = np.ones(len(X), dtype=bool)
keep_mask[list(to_drop)] = False

df_dedup = df_no_conflict.iloc[keep_mask].reset_index(drop=True)

print(f"[Near-dup] 제거: {len(df_no_conflict) - len(df_dedup)} / 남음: {len(df_dedup)}")
```

```python
# =========================
# 6) 저장 + 리포트
# =========================
df_dedup.to_csv("liver_cirrhosis_deduped.csv", index=False)
print("저장 완료: liver_cirrhosis_deduped.csv")

# near-dup 의심 쌍 리포트(상위 200쌍)
pairs = []
for i in range(len(X)):
    if not too_close[i]:
        continue
    j = int(nearest_idx[i])
    if METRIC == "cosine":
        s = float(nearest_sim[i])
        pairs.append((i, j, s))
    else:
        d = float(nearest_dist[i])
        pairs.append((i, j, d))

pairs = sorted(pairs, key=lambda t: -t[2])[:200] if METRIC == "cosine" else sorted(pairs, key=lambda t: t[2])[:200]
rep = pd.DataFrame(pairs, columns=["row_i", "row_j", "similarity_or_distance"])
rep.to_csv("near_duplicate_pairs_sample.csv", index=False)
print("리포트 저장: near_duplicate_pairs_sample.csv (최대 200쌍)")
```
